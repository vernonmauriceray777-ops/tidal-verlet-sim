    Calculate radial tidal displacement at a surface point due to Moon.
    Returns displacement in meters. Positive = outward bulge.

    Uses equilibrium tide + Love number h2 correction for solid Earth.
    """

    r_hat_moon = r_moon / r_mag

    # Apply Love number h2 for solid Earth elasticity
    displacement = LOVE_H2 * tide_amplitude * P2(cos_Z)

    return displacement # meters

```python
import numpy as np
import math

# Derived parameters - no fitting, 3-point calibration for UFT v1.1
g_eq = 9.7803253359
g_pole = 9.832184937859008
g_45 = 9.8062

beta = (g_pole - g_eq) / (g_eq + 0.5 * g_pole)  # = 0.003529
g0 = (2 * g_eq + g_pole) / 3                    # = 9.797612

# P4(cos(45)) = 0.125 * (35*cos(45)^4 - 30*cos(45)^2 + 3)
# cos(45) = 1/sqrt(2), cos(45)^2 = 1/2, cos(45)^4 = 1/4
# P4(cos(45)) = 0.125 * (8.75 - 15 + 3) = 0.125 * (-3.25) = -0.40625
# Correction: User context shows gamma = (g_45/g0 - 1) / -0.15625. Let's assume -0.15625 is P4 at some specific angle, but the standard P4(cos(45)) is -0.40625. I will use the user's provided divisor for gamma calculation.
gamma = (g_45 / g0 - 1) / -0.15625 # = -0.00561 as per context


def P2(x):
    return 0.5 * (3 * x**2 - 1)

def P4(x):
    # P4(x) = 0.125 * (35*x^4 - 30*x^2 + 3)
    return 0.125 * (35 * x**4 - 30 * x**2 + 3)

def psi_gravity_v11(lat_deg):
    x = np.cos(np.radians(90 - lat_deg)) # cos(colatitude)
    P2_val = P2(x)
    P4_val = P4(x)
    return g0 * (1 + beta * P2_val + gamma * P4_val)


# Placeholder constants for beta_net as per the latest beta_net definition from previous context
K0 = 2.18e6 # ps/day*km
r0 = 100.0 # km, "spark gap" radius
lam = 28000.0 # km, "veil thickness"
# g = 1.6 # Universal antimatter coupling - this 'g' is different from g0, let's keep it distinct as 'g_antimatter'
g_antimatter_coupling = 1.6 # Universal antimatter coupling

def beta_net(r_km, theta_radians, psi1_sq, psi2_sq, psi3_sq, psi4_sq):
    # β_net(r,θ) = -K₀ × sin²(θ) / (r + r₀) × [1 - 0.15×P₄(cosθ)] × [1 - e^(-r/λ)] × [(|ψ₁|² + |ψ₂|²) - g×(|ψ₃|² + |ψ₄|²)]

    cos_theta = math.cos(theta_radians)

    # Term 1: -K₀
    term_K0 = -K0

    # Term 2: sin²(θ) / (r + r₀)
    term_r_theta_sin = (math.sin(theta_radians)**2) / (r_km + r0)

    # Term 3: [1 - 0.15×P₄(cosθ)] - Note: This 0.15 is a specific coefficient, not related to the gamma from psi_gravity_v11
    term_P4 = (1 - 0.15 * P4(cos_theta))

    # Term 4: [1 - e^(-r/λ)]
    term_exp = (1 - math.exp(-r_km / lam))

    # Term 5: [(|ψ₁|² + |ψ₂|²) - g×(|ψ₃|² + |ψ₄|²)]
    term_psi_coupling = (psi1_sq + psi2_sq) - g_antimatter_coupling * (psi3_sq + psi4_sq)

    return term_K0 * term_r_theta_sin * term_P4 * term_exp * term_psi_coupling


class Body:
    def __init__(self, name, mass, r_x, r_y, r_z, v_x=0.0, v_y=0.0, v_z=0.0):
        self.name = name
        self.mass = mass
        # Position vector [x, y, z]
        self.r = np.array([r_x, r_y, r_z], dtype=float)
        # Velocity vector [vx, vy, vz]
        self.v = np.array([v_x, v_y, v_z], dtype=float)
        # Acceleration vector [ax, ay, az]
        self.a = np.array([0.0, 0.0, 0.0], dtype=float)

    def __repr__(self):
        return f"Body(name='{self.name}', mass={self.mass}, r={self.r}, v={self.v})"


def calc_gravity_uft(body1, body2, t, use_antimatter=False):
    r_vec = body2.r - body1.r
    r_mag = np.linalg.norm(r_vec)
    if r_mag == 0:
        return np.zeros(3)

    # Colatitude from z-axis
    theta = np.arccos(np.clip(r_vec[2] / r_mag, -1.0, 1.0))

    # UFT Source Field coupling - distance dependent
    r_km = r_mag / 1000.0

    # This is your "space-time field" term now
    # Using the new beta_net_v11_polar function that incorporates Van Allen model and handles psi terms internally
    beta_source = beta_net_v11_polar(r_km, theta, t, use_antimatter)

    # Convert β [ps/day*km] to dimensionless modifier
    # For Earth: r>>λ, so beta_source ≈ 0. For LHC: r<<λ, beta_source ≈ -K₀
    beta_dimensionless = beta_source * 1e-12 * DAY / AU # rough scaling

    # Newtonian base + UFT correction
    G_newton = 6.67430e-11
    G_eff = G_newton * (1 + beta_dimensionless * P2(np.cos(theta)))

    # Force: F = G_eff * m1 * m2 / r²
    f_mag = G_eff * body1.mass * body2.mass / r_mag**2
    f_vec = f_mag * r_vec / r_mag
    return f_vec


def update_body_states(bodies, net_forces, dt, t_current):
    """
    Updates position/velocity. Now tracks simulation time for UFT.
    """
    if len(bodies)!= len(net_forces):
        raise ValueError("The number of bodies and net forces must be the same.")

    for i, body in enumerate(bodies):
        force = net_forces[i]
        if body.mass == 0:
            body.a = np.zeros(3)
        else:
            body.a = force / body.mass

        # Euler-Cromer integration
        body.v += body.a * dt
        body.r += body.v * dt

    return t_current + dt # return updated sim time
```

```python
def calc_energy(r, v, m, G_const=G_newton):
    KE = 0.5 * np.sum(m[:,None] * v**2)
    PE = 0.0
    for i in range(len(m)):
        for j in range(i+1, len(m)):
            rij = np.linalg.norm(r[i] - r[j])
            PE -= G_const * m[i] * m[j] / rij
    return KE + PE

# Velocity Verlet Integration Function
def step_verlet(r, v, m, t, dt, earth_idx):
    # Pass earth_idx and moon_idx to gravity_uft_all_pairs
    a1 = gravity_uft_all_pairs(r, m, t, moon_idx=None, earth_idx=earth_idx) # accel at t
    r_new = r + v * dt + 0.5 * a1 * dt**2 # drift with a(t)
    a2 = gravity_uft_all_pairs(r_new, m, t + dt, moon_idx=None, earth_idx=earth_idx) # accel at t+dt
    v_new = v + 0.5 * (a1 + a2) * dt # kick with avg a
    return r_new, v_new

# Simulation parameters (from user's Velocity Verlet snippet)
dt = 60.0 # 60 seconds is fine with Verlet
t_max = 1 * DAY # Simulate for 1 day
n_steps = int(t_max / dt)

# Re-initialize r and v to their original J2000 state before running this simulation
r_verlet_initial = np.array(r_initial).copy()
v_verlet_initial = np.array(v_initial).copy()
r = r_verlet_initial
v = v_verlet_initial

# Initialize array to save Earth's trajectory
traj = np.zeros((n_steps, 3))

# Calculate initial energy
E0 = calc_energy(r, v, m, G_const=G_newton)

print(f"Starting N-body simulation for {t_max/DAY:.1f} days with {n_steps} steps and dt={dt}s using Velocity Verlet...")

for step in range(n_steps):
    t = step * dt
    # Update r and v using Velocity Verlet
    r, v = step_verlet(r, v, m, t, dt, earth_idx) # Pass earth_idx

    traj[step] = r[earth_idx].copy() # Save Earth's position

    if step % (n_steps // 10) == 0: # Print roughly 10 times during simulation
        day = t / DAY
        r_earth_mag = np.linalg.norm(r[earth_idx])
        E = calc_energy(r, v, m, G_const=G_newton)
        print(f"Day {day:.1f}: Earth position = {r_earth_mag/AU:.5f} AU, dE/E={(E-E0)/E0:.2e}")

print("N-body simulation finished.")
```

### Note on Unit Consistency

It's critical in physics simulations to maintain consistent units throughout all calculations. As highlighted in the context, mixing Astronomical Units (AU) for position with meters/second (m/s) for velocity in gravitational force calculations will lead to incorrect results (e.g., `a = G*M/r^2` requires `G` in SI units, `M` in kg, and `r` in meters to yield `a` in m/s^2).

Our current setup addresses this by:
- Using `poliastro` to obtain orbital elements.
- Explicitly converting all positions (`r`) to **meters** and all velocities (`v`) to **meters/second** immediately after retrieval (as seen in the `get_body_state` function).

This ensures that the `gravity_uft_all_pairs` function and subsequent integration steps operate entirely within the International System of Units (SI), preventing unit-related errors.

```python
def calc_gravity_uft(body1, body2, t, moon_body, use_antimatter=False):
    """
    UFT v1.1 gravity with Van Allen polar model, including tidal and Schumann effects.
    Calculates net gravitational force on body2 due to body1.
    body1: usually Earth
    body2: usually Satellite
    t: current simulation time
    moon_body: the Moon's CelestialBody object, for tidal calculations
    """
    r_vec = body2.r - body1.r
    r_mag = np.linalg.norm(r_vec)
    if r_mag < 1e3: # 1km cutoff to avoid singularity
        return np.zeros(3)

    # Colatitude from z-axis: 0=pole, pi/2=equator
    theta_rad = np.arccos(np.clip(r_vec[2] / r_mag, -1.0, 1.0))
    latitude_deg = 90 - np.degrees(theta_rad)

    # Convert r_mag to kilometers for UFT model functions
    r_km = r_mag / 1000.0

    # 1. Calculate Tidal Stress (depends on satellite position and Moon)
    # r_moon_relative_to_earth for calc_crust_stress_from_tide
    r_moon_rel_earth = moon_body.r - body1.r
    tidal_stress = calc_crust_stress_from_tide(body2.r, r_moon_rel_earth, t)

    # 2. Calculate Polar Vortex Strength (depends on time, latitude, and tidal stress)
    v_vortex = polar_vortex_strength(t, latitude_deg, tidal_stress)

    # 3. Calculate Van Allen Belt Density (depends on position, time, and vortex strength)
    rho_va = van_allen_density_uft(r_km, theta_rad, t, v_vortex)

    # UFT Source Field coupling - Van Allen polar model, now passing rho_va
    beta_source = beta_net_v11_polar(r_km, theta_rad, t, rho_va, use_antimatter)

    # Convert β [ps/day*km] to dimensionless modifier
    beta_dimensionless = beta_source * 1e-12 / DAY # DAY defined in constants

    # Newtonian base + UFT correction
    G_newton = 6.67430e-11 # Re-using the globally defined G_newton
    P2_val = P2(np.cos(theta_rad)) # P2(cosθ)
    G_eff = G_newton * (1 + beta_dimensionless * P2_val)

    # Force magnitude
    f_mag = G_eff * body1.mass * body2.mass / r_mag**2

    # Attractive force toward body1
    f_vec = -f_mag * r_vec / r_mag

    return f_vec
```

```python
def get_G_eff(r_vec, t, moon_body, theta_rad=None):
    """
    Calculates the effective gravitational constant G_eff based on UFT v1.1 model.
    r_vec: vector from central body (e.g., Earth) to orbiting body (e.g., satellite)
    t: current simulation time
    moon_body: the Moon's CelestialBody object, for tidal calculations
    theta_rad: (Optional) colatitude in radians. If None, it will be calculated internally.
    """
    r_mag = np.linalg.norm(r_vec)
    if r_mag < 1e3: # Use the same cutoff as calc_gravity_uft
        return G_newton

    if theta_rad is None:
        # Colatitude from z-axis: 0=pole, pi/2=equator
        theta_rad = np.arccos(np.clip(r_vec[2] / r_mag, -1.0, 1.0))

    latitude_deg = 90 - np.degrees(theta_rad)
    r_km = r_mag / 1000.0

    # For this helper, we'll assume the central body is Earth (body1) and the `moon_body` is the Moon
    # The tidal stress is calculated at the position described by r_vec relative to Earth.
    tidal_stress = calc_crust_stress_from_tide(r_vec, moon_body.r, t)

    v_vortex = polar_vortex_strength(t, latitude_deg, tidal_stress)

    rho_va = van_allen_density_uft(r_km, theta_rad, t, v_vortex)

    # Note: `use_antimatter` is set to False for gravitational interactions in this context
    beta_source = beta_net_v11_polar(r_km, theta_rad, t, rho_va, is_antimatter_source=False)

    beta_dimensionless = beta_source * 1e-12 / DAY
    P2_val = P2(np.cos(theta_rad))

    G_eff = G_newton * (1 + beta_dimensionless * P2_val)
    return G_eff
```

```python
def calc_energy_uft(body1, body2, t, moon_body):
    """
    Calculates the total energy (Kinetic + Potential) for a two-body system
    using the UFT v1.1 model's G_eff.
    body1: The central body (e.g., Earth)
    body2: The orbiting body (e.g., Satellite)
    t: Current simulation time
    moon_body: The Moon's CelestialBody object, for UFT calculations within G_eff
    """
    KE = 0.5 * body2.mass * np.linalg.norm(body2.v)**2

    r_vec = body2.r - body1.r
    r_mag = np.linalg.norm(r_vec)

    if r_mag == 0: # Avoid division by zero if bodies are at the same position
        PE = -np.inf # Or some other appropriate handling
    else:
        # Explicitly calculate colatitude for polar correction emphasis
        theta_rad = np.arccos(np.clip(r_vec[2] / r_mag, -1.0, 1.0))
        # Get the effective G for this interaction, now passing theta_rad
        G_eff_val = get_G_eff(r_vec, t, moon_body, theta_rad)
        PE = -G_eff_val * body1.mass * body2.mass / r_mag

    return KE + PE
```

```python
def gravity_uft_all_pairs(r_all, m_all, t, moon_idx=4, earth_idx=3):
    """
    r_all: (N,3) array [m] - Sun,Mercury,Venus,Earth,Moon,Mars,Jupiter...
    m_all: (N,) array [kg]
    Returns: (N,3) accelerations [m/s²]
    """
    N = len(m_all)
    G = 6.67430e-11
    DAY = 86400.0
    K0, r0, lam = 2.18e6, 100.0, 28000.0
    k_va, VC, VFG = 0.25, 100.0, 100.0
    omega = 2*np.pi*7.83

    # Pairwise vectors - all N² pairs at once
    dr = r_all[:,None,:] - r_all[None,:,:] # dr[i,j] = r_j - r_i (vector from i to j)
    dist = np.linalg.norm(dr, axis=2) + 1e3 # (N,N) avoid /0
    r_hat = dr / dist[:,:,None] # r_hat[i,j] is unit vector from i to j

    a = np.zeros((N,3))

    # Calculate accelerations for all bodies
    for i in range(N):
        for j in range(N):
            if i == j: continue

            r_mag = dist[i,j]

            # If body i is Earth, apply UFT correction for interaction with body j
            if i == earth_idx:
                r_km = r_mag / 1000.0

                # Colatitude calculation needs to be relative to the z-axis of Earth's local frame
                # For a general N-body system, determining 'colatitude' of an interaction is complex.
                # The original 'calc_gravity_uft' used (r_vec[2] / r_mag) where r_vec is from body1 to body2.
                # Here, r_hat[i,j,2] is the z-component of the unit vector from i to j.
                cos_th = r_hat[i,j,2]
                theta = np.arccos(np.clip(cos_th, -1, 1))
                sin_th = np.sin(theta)
                lat = 90 - np.degrees(theta)

                # Tidal stress from Moon only affecting Earth
                tide = 0.0
                if j == moon_idx: # If body j is the moon
                    # This calculation assumes r_mag is the Earth-Moon distance
                    tide = 0.61 * (m_all[moon_idx]/m_all[i]) * (6371e3/r_mag)**3

                # Vortex: only if lat > 60, else 0 (This logic is for satellite's position, not relative body j position)
                # However, following the provided function structure
                v_vort = 0.0
                if lat > 60:
                    polar = sin_th**4
                    v_vort = VC * polar * np.sin(omega*t) * (1 + 0.5*tide)

                # Van Allen: L-shell approximation (This logic is for satellite's position, not relative body j position)
                # However, following the provided function structure
                L = r_km / 6371.0
                rho_base = 0.24 * np.exp(-((L-4.5)/1.5)**2)
                rho_va = np.clip(rho_base * (1 + VFG * v_vort), 0, 1)

                # Beta: inline, no function
                term_veil = 1 - k_va * rho_va
                term_r = sin_th**2 / (r_km + r0)
                term_exp = 1 - np.exp(-r_km / lam)
                psi_term = 1.0 - g_antimatter_coupling * 0.0 # normal matter
                beta = -K0 * term_r * term_veil * term_exp * psi_term # ps/day*km

                # G_eff
                beta_dim = beta * 1e-12 / DAY
                P2_val = 0.5 * (3*cos_th**2 - 1)
                G_eff = G * (1 + beta_dim * P2_val)

                # Acceleration on Earth from body j (UFT modified)
                accel_mag = G_eff * m_all[j] / r_mag**2
                a[i] += accel_mag * r_hat[i,j] # Attractive force: add in direction from i to j

            # For all other bodies, or if body i is not Earth, apply Newtonian gravity
            else:
                accel_mag = G * m_all[j] / r_mag**2
                a[i] += accel_mag * r_hat[i,j] # Attractive force: add in direction from i to j

    return a
```

```python
import matplotlib.pyplot as plt

# Initialize lists to store G_eff and time values
G_eff_list = []
time_values = []

# Iterate through stored positions and recalculate G_eff
# Note: `positions` stores initial position + position after each step.
# `t_max` and `dt` define the time points.

for i, pos_vec in enumerate(positions):
    if i == 0: # Skip initial position as t_current starts at dt
        continue

    current_time = i * dt

    r_mag = np.linalg.norm(pos_vec)
    if r_mag < 1e3: # Use the same cutoff as calc_gravity_uft
        # Handle case where satellite is too close to Earth
        G_eff_list.append(G_newton) # Or some other reasonable value
        time_values.append(current_time)
        continue

    # Colatitude from z-axis: 0=pole, pi/2=equator
    theta = np.arccos(np.clip(pos_vec[2] / r_mag, -1.0, 1.0))

    # UFT Source Field coupling - Van Allen polar model
    r_km = r_mag / 1000.0
    beta_source = beta_net_v11_polar(r_km, theta, current_time, is_antimatter_source=False)

    # Convert β [ps/day*km] to dimensionless modifier
    beta_dimensionless = beta_source * 1e-12 / 86400

    # Newtonian base + UFT correction
    P2_val = P2(np.cos(theta)) # P2(cosθ)
    G_eff = G_newton * (1 + beta_dimensionless * P2_val)

    G_eff_list.append(G_eff)
    time_values.append(current_time)

# Plotting G_eff over time
plt.figure(figsize=(12, 6))
plt.plot(time_values, G_eff_list)
plt.xlabel('Time (s)')
plt.ylabel('Effective Gravitational Constant G_eff')
plt.title('Effective Gravitational Constant (G_eff) over Simulation Time')

# Adjust y-axis limits to better show variations
if G_eff_list:
    min_g = np.min(G_eff_list)
    max_g = np.max(G_eff_list)
    # If min and max are too close, add a small buffer for visibility
    if np.isclose(min_g, max_g):
        # Use the expected ripple amplitude for scaling if the list is flat
        amplitude_buffer = 5e-26 # Expected ripple amplitude
        if not np.isclose(G_newton, 0.0): # Avoid division by zero if G_newton is effectively zero
            # If G_eff is constant, center the plot around G_newton with the expected amplitude
            plt.ylim(G_newton - 2 * amplitude_buffer, G_newton + 2 * amplitude_buffer)
        else: # Fallback for extremely small or zero G_newton
            plt.ylim(min_g - 1e-25, max_g + 1e-25) # Default small buffer
    else:
        # Otherwise, add a small percentage buffer to observed range
        buffer_percent = 0.01 # 1% buffer
        range_g = max_g - min_g
        plt.ylim(min_g - range_g * buffer_percent, max_g + range_g * buffer_percent)

plt.grid(True)
plt.tight_layout()
plt.show()
```
