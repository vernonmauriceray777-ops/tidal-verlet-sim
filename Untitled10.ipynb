```python
# Physical Constants
G = 6.674e-11 # Gravitational constant (N(m/kg)^2)
c = 299792458.0 # Speed of light (m/s)

# Black Hole (Central Body) Parameters
M_central = 1.989e30 # Mass of the central body (e.g., Sun's mass in kg)
J_bh = G * M_central**2 / c # Black hole angular momentum (derived from a=0.9 for Kerr BH) ~2e53
# J_bh = 1e67 # User-specified adjustable value for J_bh

# Orbital Parameters
r_orbit = 1e8 # Initial orbital radius (meters)
dt = 1e-3 # Time step for integration (seconds)
num_steps = 100000 # Number of simulation steps

# Simulation Flags
USE_FRAME_DRAG = True # Set to True to enable frame-dragging force, False for Newtonian only
```

```python
import numpy as np

def frame_dragging_force(r_vec, v_vec, M_bh, J_bh):
    """Calculates the Lense-Thirring (frame-dragging) force per unit mass.

    This function implements the Lense-Thirring force formula, a relativistic
    effect that causes frame-dragging around a rotating mass (black hole).
    The force is calculated based on the position and velocity vectors of
    the orbiting particle, the mass of the central body, and its angular
    momentum.

    Args:
        r_vec (np.ndarray): Position vector of the orbiting particle relative
                            to the central body (3D array).
        v_vec (np.ndarray): Velocity vector of the orbiting particle (3D array).
        M_bh (float): Mass of the central black hole (kg).
        J_bh (float): Angular momentum of the central black hole (kg m^2/s).

    Returns:
        np.ndarray: The frame-dragging force per unit mass (acceleration)
                    acting on the particle (3D array). Returns a zero vector
                    if the particle is too close to the singularity.
    """
    # G and c are now assumed to be globally defined in cell 2ebe827a

    r = np.linalg.norm(r_vec)

    if r < 1e4: # avoid singularity at r=0
        return np.zeros(3)

    # J along z-axis
    J_vec = np.array([0, 0, J_bh])

    # Pre-calculate cross products and dot products
    r_cross_v = np.cross(r_vec, v_vec)
    J_cross_v = np.cross(J_vec, v_vec)
    J_dot_r = np.dot(J_vec, r_vec)

    # Total Lense-Thirring force factor (excluding particle mass 'm' as it's handled in calculate_forces)
    # The formula F = (2G/c^2) * m * [...] means the factor is 2G/c^2
    prefactor = (2 * G) / (c**2 * r**5)
    F_fd = prefactor * (J_cross_v * r**2 - 3 * J_dot_r * r_cross_v)

    # The force returned here is per unit mass of the orbiting particle
    return F_fd
```

```python
def calculate_forces(pos, vel, masses):
    """Calculates the gravitational forces and optionally frame-dragging force on all particles.

    Args:
        pos (np.ndarray): 3D array of particle positions (N, 3).
        vel (np.ndarray): 3D array of particle velocities (N, 3).
        masses (np.ndarray): 1D array of particle masses (N,).

    Returns:
        np.ndarray: 3D array of forces (N, 3) acting on each particle.
    """
    N = len(masses)
    forces = np.zeros((N, 3))

    # Newtonian Gravitational Force
    # Simplified for a central body and orbiting particles.
    # Only calculate force on orbiting particles (index 1 onwards) from central body (index 0)
    for i in range(1, N):
        r_vec = pos[i] - pos[0]
        r_magnitude = np.linalg.norm(r_vec)
        if r_magnitude == 0:
            continue

        # F = G * m1 * m2 / r^2
        grav_force_magnitude = (G * masses[0] * masses[i]) / r_magnitude**2
        grav_force_vec = -grav_force_magnitude * (r_vec / r_magnitude)
        forces[i] += grav_force_vec

    # Add Lense-Thirring (frame-dragging) force if enabled
    if USE_FRAME_DRAG:
        # The frame-dragging_force function returns force per unit mass (acceleration)
        # So we multiply by the particle's mass to get the actual force.
        for i in range(1, N):
            # Note: frame_dragging_force assumes J_bh is globally defined
            fd_acceleration = frame_dragging_force(pos[i] - pos[0], vel[i], masses[0], J_bh)
            forces[i] += masses[i] * fd_acceleration

    return forces
```

```python
# Initial setup of positions, velocities, and masses
# Schwarzschild radius (needed for v_circ calculation consistency)
r_s = (2 * G * M_central) / c**2 # Re-calculate for robustness
v_circ = np.sqrt(G * M_central * r_orbit / (r_orbit - r_s)**2)

demo_pos = np.array([
    [0,0,0],                    # BH at origin
    [r_orbit,0,0],              # Particle A: equatorial orbit in XY plane
    [0,r_orbit,0],              # Particle B: equatorial orbit (90 deg phase shift) in XY plane
    [0,0,r_orbit]               # Particle C: polar orbit in XZ plane
])
demo_vel = np.array([
    [0,0,0],                    # BH is stationary
    [0,v_circ,0],               # Particle A: velocity in +Y direction
    [-v_circ,0,0],              # Particle B: velocity in -X direction
    [v_circ,0,0]                # Particle C: velocity in +X direction
])

# Particle masses: Central body (M_central) and three orbiting particles (e.g., Earth's mass)
masses = np.array([M_central, 5.972e24, 5.972e24, 5.972e24])

# Initialize pos and vel for the main simulation run
pos = demo_pos.copy()
vel = demo_vel.copy()

saved_pos = []
saved_vel = []

# Perform Leapfrog Integration for the main simulation
# First half-step velocity kick
forces = calculate_forces(pos, vel, masses)
forces[0] = np.zeros(3) # Central body is fixed
vel += 0.5 * forces * dt / masses[:,None]

for step in range(num_steps):
    # Full drift step
    pos += vel * dt

    # Full kick step
    forces = calculate_forces(pos, vel, masses)
    forces[0] = np.zeros(3) # Central body is fixed
    vel += forces * dt / masses[:,None]

    # Save positions and velocities at intervals
    if step % 100 == 0:
        saved_pos.append(pos.copy())
        saved_vel.append(vel.copy())

# Final half-kick to synchronize velocity with position at the end of the simulation
forces = calculate_forces(pos, vel, masses)
forces[0] = np.zeros(3)
vel += 0.5 * forces * dt / masses[:,None]

# Convert saved lists to NumPy arrays
saved_pos = np.array(saved_pos)
saved_vel = np.array(saved_vel)

print("Initial simulation setup and run complete. saved_pos and saved_vel are populated.")
```
